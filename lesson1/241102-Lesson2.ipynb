```python
#讓使用者輸入被除數(整數)及除數(整數，不可以是零)
#程式會顯示兩數相除的商及餘數。
a1=int(input("輸入被除數(整數)"))
a2=int(input("除數(整數，不可以是零)"))
print("商:",a1//a2)
print("餘:",a1%a2)
```

```python
print(1,2.2,'33')
print(1,2.2,'33',sep='~',end='.\n')
#        ↑引數    ↑引數值
```

```python
print(a1//a2,a1%a2,sep=",")
```
