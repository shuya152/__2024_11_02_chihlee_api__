```python
from openpyxl import load_workbook, Workbook, worksheet

wb = load_workbook('aqi.xlsx')
type(wb)
```

```python
sheet:worksheet = wb.worksheets[0]
type(sheet)
sheets:list=[]
column_names:list = [ cell.value for cell in list(sheet)[0] ]
print(column_names)
```

```python
# test
for row in list(sheet)[1:]:
    column_name = [cell.value for cell in row]
    print(column_name)
```

```python
# test
for row in list(sheet)[1:]:
    for idx, item in enumerate(row):
        print(column_names[idx], item.value)
        print('====================================')
```

```python
for row in list(sheet)[1:]:
    site:dict = {}
    for idx, item in enumerate(row):
        site[column_names[idx]] = item.value
    sheets.append(site)

print(len(sheets))
print(sheets)
```
